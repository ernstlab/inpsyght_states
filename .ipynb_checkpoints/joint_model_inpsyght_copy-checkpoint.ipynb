### Annotation tests

```python
def marginal_model(df1, df2, feature_cols, meta_cols, scaled=True):
    odds = []
    ps = []
    a_meta = df1[meta_cols].values
    b_meta = df2[meta_cols].values
    endog = [1] * len(a_meta) + [0] * len(b_meta)
    for c in feature_cols:
        a = list(df1[c])
        b = list(df2[c])
        x = np.concatenate((np.asarray(a), np.asarray(b)), axis=0).reshape(-1, 1)
        x_meta = np.concatenate((np.asarray(a_meta), np.asarray(b_meta)), axis=0)
        if scaled:
            exog = scale(np.concatenate((x, x_meta), axis=1), axis=0)
        else:
            exog = np.concatenate((x, x_meta), axis=1)
        exog = sm.add_constant(exog)
        logit = Logit(endog, exog)
        try:
            results = logit.fit(disp=False, maxiter=1000)
        except np.linalg.LinAlgError as err:            
            ps.append(-1)
            odds.append(0)
            continue
        ps.append(results.pvalues[1])
        odds.append(results.params[1])
    return odds, ps
```

```python
def annotation_test(wgspd_df, cases, controls, features, covariates):
    if cases == "BP":
        df1 = wgspd_df[wgspd_df["Pheno"] == "BP"]
    elif cases == "SCZ":
        df1 = wgspd_df[wgspd_df["Pheno"] == "SCZ"]
    elif cases == "BOTH":
        df1 = wgspd_df[(wgspd_df["Pheno"] == "SCZ") | (wgspd_df["Pheno"] == "BP")]
    elif cases == "TOPMED":
        df1 = wgspd_df[(wgspd_df["Pheno"] == "CTRL") & (wgspd_df["Group"] == "TOPMED")]
    
    if controls == "INPSYGHT":
        df2 = wgspd_df[(wgspd_df["Pheno"] == "CTRL") & (wgspd_df["Group"] == "InPSYght")]
    elif controls == "BOTH":
        df2 = wgspd_df[wgspd_df["Pheno"] == "CTRL"]
            
    odds, ps = marginal_model(df1, df2, features, covariates, scaled=True)
    odds_df = pd.DataFrame()
    odds_df["annotation"] = [item for item in features]
    odds_df.index = odds_df["annotation"]
    odds_df["logOR"] = odds
    odds_df["p"] = ps
    odds_df["-logP"] = -1 * np.log10(ps)
    odds_df.index = [item.split('/')[-1].replace(".bed.numeric", "") for item in odds_df["annotation"]]
    odds_df["desc"] = state_descs["desc"]
    
    types = []
    for item in odds_df.index:
        if item[0] == "E" and item[1:].isdigit():
            types.append("ChromHMM state")
        elif item[0] == "U" and item[1:].isdigit():
            types.append("ConsHMM state")
        else:
            types.append("Repeats")
    odds_df["Annotation"] = types
    
    return odds_df
```

```python
cases = ["TOPMED", "BOTH", "BP", "SCZ", "BOTH", "BP", "SCZ"]
controls = ["INPSYGHT", "BOTH", "BOTH", "BOTH", "INPSYGHT", "INPSYGHT", "INPSYGHT"]
titles = ["TOPMed controls vs. InPSYght controls", "BD + SZ vs. InPSYght + TOPMed controls", "BD vs. InPSYght + TOPMed controls", 
          "SZ vs. InPSYght + TOPMed controls", "BD + SZ vs. InPSYght controls", "BD vs. InPSYght controls", 
          "SZ vs. InPSYght controls"]
```

```python
covariates1 = ["newC%d" % i for i in range(1, 11)] + ["Sex", "Total_snps"]  + ["meta_PC_new%d" % i for i in range(1, 5)] 
covariates2 = ["newC%d" % i for i in range(1, 11)] + ["Sex", "Total_snps"]
```

```python
odds_dfs = []
odds_dfs_nometa = []
for i in range(len(cases)):
    c = covariates1
    if controls[i] == "INPSYGHT" and cases[i] != "TOPMED":
        c = covariates1 + ["wave_%d.0" % i for i in range(6)]
    odds_df = annotation_test(wgspd_df, cases[i], controls[i], feature_cols, c)
    odds_df["mean %"] = wgspd_df_portions[feature_cols].mean(axis=0) * 100
    odds_df["abs logOR"] = np.abs(odds_df["logOR"])
    odds_dfs.append(odds_df)
    
    c = covariates2
    if controls[i] == "INPSYGHT" and cases[i] != "TOPMED":
        c = covariates2 + ["wave_%d.0" % i for i in range(6)]
    odds_df = annotation_test(wgspd_df, cases[i], controls[i], feature_cols, c)
    odds_df["mean %"] = wgspd_df_portions[feature_cols].mean(axis=0) * 100
    odds_df["abs logOR"] = np.abs(odds_df["logOR"])
    odds_dfs_nometa.append(odds_df)
    
```

```python
def qqplot(df, name, label="", color="Blue", linewidth=0):
    observed = np.sort(-1 * np.log10(df["p"]))
    expected = np.sort(-1 * np.log10(np.arange(1, len(df) + 1) / len(df)))
    # plt.figure(figsize=(7, 7))
    expected_d = chi2.ppf(0.5, df=1)
    observed_d = chi2.ppf(1 - np.median(df["p"]), df=1)
    l = observed_d / expected_d
    sns.scatterplot(x=expected, y=observed, linewidth=linewidth, 
                    label=label + " $\lambda$=%.2f" % l, color=color, edgecolor="black")
    plt.axline([0, 0], [1, 1], ls='-', color="#555555")
    plt.xlabel("Expected $-log_{10}(p)$", size=14)
    plt.xticks(fontsize=12)
    plt.ylabel("Observed $-log_{10}(p)$", size=14)
    plt.yticks(fontsize=12)
    plt.title(name, size=15)
    return l
```
